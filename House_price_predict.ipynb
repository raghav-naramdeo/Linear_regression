```python
df3.total_sqft.unique()
# here we have data as ranges (for eg : '1133 - 1384')
```

```python
# converting ranges into single data.
# checking for the data in range format.
def is_float(x):
  try:
    float(x)
  except:
    return False
  return True
```

```python
#"~" is a negate function, it will hreturn only data, that is in ranges.
df3[~df3['total_sqft'].apply(is_float)].head(10)
```

```python
# only working with the range values, we will ignore other data(eg :34.46Sq. Meter,4125Perch	 )
# creating a function to take average of ranges.
def convert_sqft_to_num(x):
  tokens = x.split('-')
  if len(tokens) ==2:
    return(float(tokens[0])+float(tokens[1]))/2
  try:
    return float(x)
  except:
    return None
```

```python
convert_sqft_to_num("4125Perch")
```

```python
df4 = df3.copy()
df4['total_sqft'] = df4['total_sqft'].apply(convert_sqft_to_num)
df4.head(5)
```

```python
df5 = df4.copy()
df5['price_per_sqft'] = df5['price']*100000/df5['total_sqft']
df5.head()
```

```python
#dealing with location, to many data
len(df5.location.unique())
```

```python
#clean leading and trailing space
df5.location = df5.location.apply(lambda x :str(x).strip())
location_stats = df5.groupby('location')['location'].agg('count').sort_values(ascending=False)
location_stats
```

```python
# categorizing as other location.
len(location_stats[location_stats<=10])
```

```python
location_stats_less_than_10 = location_stats[location_stats<=10]
```

```python
df5.location = df5.location.apply(lambda x : 'other' if x in location_stats_less_than_10 else x)
len(df5.location.unique())
```

```python
#outlier(anomalies) :- detection and removal anomalies
df5[df5.total_sqft/df5.bhk<300].head() #house with bedroom area less than 300sqft
```

```python
df5.shape
```

```python
#Negate outliers
df6 = df5[~(df5.total_sqft/df5.bhk<300)]
df6.shape
```

```python
# checking price per sqft
df6.price_per_sqft.describe()
```

```python
# function to remove extreme price using Standard deviation.
#removing data greater than 1 SD per location
def remove_pps_outliers(df):
  df_out = pd.DataFrame()
  for key, subdf in df.groupby('location'):
    m = np.mean(subdf.price_per_sqft)
    sd= np.std(subdf.price_per_sqft)
    reduced_df = subdf[(subdf.price_per_sqft > (m-sd)) & (subdf.price_per_sqft <= (m+sd))]
    df_out = pd.concat([df_out,reduced_df],ignore_index=True)
  return df_out

df7 = remove_pps_outliers(df6)
df7.shape
```

```python
# Visualizing the data
def plot_scatter_chart(df, location):
  bhk2 = df[(df.location == location) & (df.bhk == 2)]
  bhk3 = df[(df.location == location) & (df.bhk == 3)]
  plt.rcParams['figure.figsize']=(15,10)
  plt.scatter(bhk2.total_sqft,bhk2.price,color ='blue',label ='2 BHK', s=50)
  plt.scatter(bhk3.total_sqft,bhk3.price,marker='+',color ='green',label ='3 BHK', s=50)
  plt.xlabel("Total squar feet area")
  plt.ylabel("price")
  plt.title(location)
  plt.legend()

plot_scatter_chart(df7,"Hebbal" )
```

```python
def remove_bhk_outliers(df):
  exclude_indices = np.array([])
  for location, location_df in df.groupby('location'):
    bhk_stats = {}
    for bhk, bhk_df in location_df.groupby('bhk'):
      bhk_stats[bhk]={
          'mean' : np.mean(bhk_df.price_per_sqft),
          'std' : np.std(bhk_df.price_per_sqft),
          'count': bhk_df.shape[0]
      }
    for bhk, bhk_df in location_df.groupby('bhk'):
      stats = bhk_stats.get(bhk-1)
      if stats and stats['count'] >5:
        exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_sqft < (stats['mean'])].index.values)
  return df.drop(exclude_indices,axis='index')

df8 =remove_bhk_outliers(df7)
df8.shape
```

```python
df9 = df8[df8.bath<df8.bhk+2]
df9.shape
```

```python
df10 = df9.drop(['size','price_per_sqft'],axis='columns')
df10.head(4)
```

**MAKING ML MODEL**

```python
# converting location string data into num vals by one-hot encoding (dummies)
dummies = pd.get_dummies(df10.location)
dummies = dummies.astype(int) # bool value to binary
dummies.head(3)
```

```python
# dropping one column(others) to avoid dummy variable trap(avoid multicolinaerity)
df11 = pd.concat([df10,dummies.drop('other',axis='columns')],axis ='columns')
df12 = df11.drop('location',axis='columns')
df12.head(3)
```

```python
df12.shape
```

```python
#marking the independent variable as x and dependent as y
x =df12.drop('price',axis='columns')
x.head()
```

```python
y = df12.price
y.head()
```

```python
#dividing train and test data set in 80-20, using sklearn model selection

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=10)
```

```python
# creating a linear regression model using sklearn linear model

from sklearn.linear_model import LinearRegression
# Model
lr_clf = LinearRegression()
#Train
lr_clf.fit(x_train,y_train)
#Evaluate
lr_clf.score(x_test,y_test)
```

```python
#using K fold cross validation

from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

#Randomise the sample for each fold
cv = ShuffleSplit(n_splits = 5, test_size = 0.2, random_state = 0)


cross_val_score(LinearRegression(), x, y, cv=cv)
```

```python
# trying different regression, using grid search CV (API of sklearn)

from sklearn.model_selection import GridSearchCV

from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor

def find_best_model_using_gridsearchcv(x,y):
  algos = {
      'linear_regression': {
          'model' : LinearRegression(),
          'params': {
              'fit_intercept' : [True,False]
          }
      },
      'lasso' : {
          'model' : Lasso(),
          'params': {
              'alpha' : [1,2],
              'selection' :['random', 'cyclic']
          }
      },
      'decision_tree' : {
          'model' : DecisionTreeRegressor(),
          'params': {
              'criterion': ['friedman_mse', 'squared_error'],
              'splitter' : ['best','random']
          }
      }
  }
  scores =[]
  cv = ShuffleSplit(n_splits = 5, test_size=0.2,random_state=0)
  for algo_name, config in algos.items():
    gs = GridSearchCV(config['model'], config['params'], cv = cv, return_train_score=False)
    gs.fit(x,y)
    scores.append({
        'model' : algo_name,
        'best_score' : gs.best_score_,
        'best_params' : gs.best_params_
    })
  return pd.DataFrame(scores, columns=['model','best_score','best_params'])
find_best_model_using_gridsearchcv(x,y)
```

```python
 find_best_model_using_gridsearchcv(x,y)

 #Linear regression is the best one
```

```python
#So we will use our linear trained model to predict data

def predict_price(location,sqft,bath,bhk):
  loc_index = np.where(x.columns == location)[0][0]

  z = np.zeros(len(x.columns))
  z[0] = sqft
  z[1] = bath
  z[2] = bhk
  if loc_index >= 0:
    z[loc_index] = 1

  return lr_clf.predict([z])[0]
```

```python
#Example usage
predict_price('1st Phase JP Nagar', 1000,2,2)
```
